```python
# looking at the data set
data_source_file = "https://covidtracking.com/data/download/new-york-history.csv"
df_case = pd.read_csv(data_source_file).set_index('date')
df_case = df_case.loc[:,'positive']
fig, cases = pit.subplots()
cases.bar(df_case.index, df_case.values)
cases.xaxis.set_major_locator(pit.MaxNLocator(15))
pit.xticks(rotation=45)
```
